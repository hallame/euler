```python
def dfs(u, graph, visited_edge, path=[]):
    path = path + [u]

    # Поиск пути
    # v - текущая вершина. u - вершина которую предлагается посетить
    for v in graph[u]: #перебираем u по смежным вершинам
        if visited_edge[u][v] == False:  # если по ребру еще не было прохода
            visited_edge[u][v], visited_edge[v][u] = True, True  # делаем проход по ребру (отмечаем в обе стороны)
            path = dfs(v, graph, visited_edge, path) # добавляем ребро в путь
    return path


# Проверка наличия в графе эйлерова цикла или пути на основе четности степеней вершин
def check_circuit_or_path(graph, max_node):
    odd_degree_nodes = 0
    odd_node = -1
    for i in range(max_node):
        if i not in graph.keys():
            continue
        if len(graph[i]) % 2 == 1:  # проверка четности
            odd_degree_nodes += 1
            odd_node = i
    print(f"содержатся узлы с НЕчетными степенями:{odd_degree_nodes}")
    if odd_degree_nodes == 0:
        return 1, odd_node
    if odd_degree_nodes == 2:
        return 2, odd_node
    return 3, odd_node


def check_euler(graph, max_node):
    visited_edge = [[False for _ in range(max_node + 1)] for _ in range(max_node + 1)]
    check, odd_node = check_circuit_or_path(graph, max_node)
    
    if check == 3:
        print("В графе нет цикла Эйлера")
        return
    
    start_node = 32
    if check == 2:
        start_node = odd_node
        print("В графе нет цикла Эйлера, но есть путь Эйлера")
        
    if check == 1:
        print("Граф является Эйлеровым")
        
    path = dfs(start_node, graph, visited_edge)
    
    print(path)
```
